## Selecting data from dataframe

```python
df['species'] #select 'species' column
```

```python
df['species'].value_counts() # return count for each unique value
```

```python
df['species'].nunique() # returns the number of unique values
```

* .iloc interger based selection
* .loc  index-name selection

```python
df.iloc[0:5,:] #row, col
# equivalent to df.head() returns first 5 rows and all columns
```

```python
df.iloc[:,0:3] # returns all rows and first 3 columns
```

```python
df.loc[:,'petal_length':'species'] #retuns specified column range all rows
```

```python
petal_data = df.loc[:, 'petal_length' : 'petal_width']
petal_data.head()
```

```python
sepal_data = df.loc[:, 'sepal_length':'sepal_width']
sepal_data.head()
```

```python
target_data = df['species']
target_data.head()
```

```python
# merge data sets w/ common index
msmt_data = pd.merge(sepal_data, petal_data, left_index = True, right_index = True)
msmt_data.head()
```

```python
all_data = pd.merge(msmt_data, target_data, left_index = True, right_index= True)
all_data.head()
```

```python
all(df == all_data) # data frames are equal after merging data components
```
